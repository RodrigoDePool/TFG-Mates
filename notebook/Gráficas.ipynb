```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def cantor_set(s,f,depth):
    if depth==1:
        return [s,f]
    step = (f-s)/3
    left = cantor_set(s, s+step, depth-1)
    right = cantor_set(f-step, f, depth-1)
    return left + right 

x = cantor_set(0,1,depth=4)
y = [0]*len(x)
```

```python
x
```

```python
def center_and_radius(n,m):
    s = (n-1/3)/3**m
    f = (n+4/3)/3**m
    return ((s+f)/2, (s+f)/2 - s)

circles = []
m=1
n = [0,2]
for ele in n:
    circles.append(center_and_radius(ele,m))

m=2
n = [0,2,6,8]
for ele in n:
    circles.append(center_and_radius(ele,m))

m=3
n = [0,2,6,8,18,20,24,26]
for ele in n:
    circles.append(center_and_radius(ele,m))
```

```python
fig,ax = plt.subplots()
for c,r in circles:
    circ=plt.Circle((c,0),r,fill=False)
    ax.add_artist(circ)
ax.set_aspect(1)
plt.plot([-10,199],[0,0], linewidth=0.1, color='black')
plt.xlim(-1/7,1+1/7)
plt.ylim(-1/3,1/3)
plt.scatter(x,y, s=[5]*len(x), color='r')

plt.savefig('conjuntoD.png')
```
